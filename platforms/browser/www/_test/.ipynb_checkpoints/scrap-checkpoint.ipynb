```python
# Importing modules
import pandas as pd
```

```python
# Reading data

excel_file = 'HSK1-3.xlsx'
df = pd.read_excel(excel_file)
```

```python
# Editing data
chinese = []
test = ["question","question","question"]

for index, row in df.iterrows():
    _id = index + 1
    character = row['Chinese']
    pinay = row['Pinyin']
    meaning = row['English']
    
    dic = {"id": _id,"character": character, "pinay": pinay, "meaning": meaning, "test": test}
    chinese.append(dic)
    
#finale_json = {"chinese" : chinese}
#print(finale_json)
print(chinese)
```

```python
# Saving as JSON
df.to_json("testfile")
```
